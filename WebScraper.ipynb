# Immportando libereias

```python
import requests
from bs4 import BeautifulSoup #>>Parsear el codigo HTML
```

```python
#Guardando url
url = 'https://www.pagina12.com.ar/'
```

# Haciendo la solicitud

```python
p = requests.get(url)
```

```python
#Checando el status
p.status_code
```

```python
print(p.text)
```

```python
#Encabexado de la respuest
p.headers
```

```python
#Encabezado de la solicitud
p.request.headers
```

```python
#Ver el metodo utilizado
p.request.method
```

```python
p.request.url
```

```python
s = BeautifulSoup(p.text, 'html.parser')
```

```python
type(s)
```

## Ver como esta estructurada la pagina

```python
print(s.prettify())
```

```python
#Etiqueta que contiene las secciones
"""
find: Devuelve el primer elemento con la estructura dada.
"""
s.find('ul')
```

```python
#Seleccionando una etiqueta en especifico
s.find('ul', attrs={'class':'horizontal-list main-sections hide-on-dropdown'})
```

```python
#Buscando los li de la seccion seleccionada
s.find('ul', attrs={'class':'horizontal-list main-sections hide-on-dropdown'}).find_all('li')
```

```python
seccions = s.find('ul', attrs={'class':'horizontal-list main-sections hide-on-dropdown'}).find_all('li')
```

```python
#Tomando el primer elemento
seccion = seccions[0]
```

```python
#Obtener en forma de texto del atributo href
seccion.a.get('href')
```

```python
seccion.get_text()
```

```python
#Obtenemos las secciones de cada link
seccions_links = [seccion.a.get('href') for seccion in seccions]
```

```python
#Links de cada seccion
seccions_links
```

```python
#Extrallendo el link de la primera seccion
#Haciendo get del primer link
#Solicitud al primer link

sec = requests.get(seccions_links[0])
```

```python
sec.status_code
```

```python
"""Parsenado con BeautifulSoup"""

s_seccion = BeautifulSoup(sec.text, 'html.parser')
```

```python
featured_article = s_seccion.find('div', attrs={'class': 'articles-list'})
featured_article
```

```python
"""Obteninendo link"""
featured_article.a.get('href')
```

```python
article_list = s_seccion.find('div', attrs={'class':'article-item__content'})
```
